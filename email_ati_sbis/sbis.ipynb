```python
import requests
import re
import pandas as pd
from tqdm import tqdm 
import time
```

```python
ati_rus: dict[Hashable, Any] = pd.read_excel(r"D:\Job\Prometeo\ati.su\xlsx_files\ati_rus_04.12.2023.xlsx").to_dict('index')
```

```python
ati_rus_lst = list(ati_rus.values())
```

```python
ati_rus_lst[17]
```

```python
def execution_time(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        func(*args, **kwargs)
        end_time = time.time()
        print("Execution Time: {:.3f} s".format(end_time - start_time))

    return wrapper
```

```python
start_time = time.time()

pbar = tqdm(total=len(ati_rus_lst))
base_url = "https://sbis.ru/contragents/{}"
for i in range(len(ati_rus_lst)):
    # if i == 15:
    #     break
    
    email_pattern = r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,4}\b"

    
    res = requests.get(base_url.format(int(ati_rus_lst[i].get("INN"))))
    ati_rus_lst[i]["Email Adresses"] = ", ".join(set(re.findall(email_pattern, str(res.text))))
    pbar.update()

pbar.close()

end_time = time.time()
print("Execution Time: {:.3f} s".format(end_time - start_time))
```
